# scRNA-Seq Analysis of DLBCL_Tumor_Cell_States

```python
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
import seaborn as sns
import dask.dataframe as dd

print("All libraries are successfully imported")
```

```python
# verbosity: errors (0), warnings (1), info (2), hints (3)
sc.settings.verbosity = 3
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor="white")
```

```python
# Load the raw count matrix
adata_matrix = sc.read_text("data/raw/GSE182434_raw_count_matrix.txt")
# Load the annotation file
annotations = pd.read_csv("data/raw/GSE182434_cell_annotation.txt", sep="\t")
```

```python
print(adata_matrix)
#print(annotations)
```

```python
# Transpose data
adata = adata_matrix.transpose()
adata
```

```python

# Add the annotations to the AnnData object
adata.obs = annotations
adata.obs
adata.var
adata.to_df()
```

## Quality Control

```python
# Mitochondrial genes, "MT-"
adata.var["mt"] = adata.var_names.str.startswith("MT-")
# Ribosomal genes
adata.var["ribo"] = adata.var_names.str.startswith("RPS", "RPL")
# Haemoglobin
adata.var["hb"] = adata.var_names.str.contains("^HB[^(P)]")
```

```python
sc.pp.calculate_qc_metrics(adata, qc_vars=["mt", "ribo", "hb"], inplace=True, log1p=True)
```

```python
# Violin plots of some of the computed quality measures:
# The number of genes expressed in the count matrix
# The total counts per cell
# The percentage of counts in mitochondrial genes

sc.pl.violin(adata, ["n_genes_by_counts", "total_counts", "pct_counts_mt"], jitter=0.4, multi_panel=True)
```

## Preprocessing

```python
# Show genes that yield the highest fraction of counts
sc.pl.highest_expr_genes(adata, n_top=20)
```

```python
# Basic Filtering
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
```
